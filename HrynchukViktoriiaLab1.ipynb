Завдання 2: Розв’язати квадратне рівняння $ax^2+bx + c = 0$.  Напишіть функцію,яка приймає параметри квадратного рівняння, а повертає одне рішення або декілька.

Наприклад:  


```
solve_quad(a=1, b=-2, c=1) -> 1  
solve_quad(a=1, b=0, c=-1) -> (-1, 1)  
solve_quad(a=0, b=2, c=-1) -> 0.5
```







````python
def solve_quad(a: float, b: float, c: float):
      discriminant = b**2 - 4*a*c
      if discriminant > 0:
         return ((-b + discriminant**0.5) / (2*a), (-b - discriminant**0.5) / (2*a))
      elif discriminant == 0:
         return -b / (2*a)
      else:
         return None
       
````

Завдання 8: Написати функцію яка приймає на вхід двовимірний масив, а повертає масив з діагональними елементами цієї матриці.

Наприклад: `[[1, 2], [3, 4]] -> [1 4]`

````python
def get_diag(X):
    return np.diagonal(X)

print(get_diag([[1,2,3],
                [3,1,5]]))


````

Завдання 9: Написати функцію яка приймає на вхід двомірний масив, а повертає масив елементів які знаходяться на другорядній діагоналі. Приклад: `[[1, 2], [3, 4]] -> [2 3]`

````python
def sec_diag(X):
    return np.fliplr(X).diagonal()

print(sec_diag([[1,2,3],
                [3,1,5]]))
````

Завдання 10: Кого більше - чоловіків чи жінок?

````python

count_of_male = dataset['Sex'].value_counts().get('male',0)
count_of_female = dataset['Sex'].value_counts().get('female',0)


print(count_of_male)

if count_of_male > count_of_female: 
    print('there are more man', count_of_male) 
else:
   print('there are more women',count_of_female) 
````

Завдання 11: Виведіть наступні характеристики пасажирів: середній вік пасажирів, середній вік чоловіків, середній вік жінок, максимальний і мінімальний вік серед чоловіків і жінок.

````python
male_dataset = dataset[(dataset['Sex'] == 'male')]
female_dataset = dataset[(dataset['Sex'] == 'female')]


print('середній вік чоловіків',male_dataset['Age'].mean())
print('мінімальний вік серед чоловіків',male_dataset['Age'].min())
print('максимальний вік серед чоловіків',male_dataset['Age'].max())

print('середній вік жінок',female_dataset['Age'].mean())
print('мінімальний вік серед жінок',female_dataset['Age'].min())
print('максимальний вік серед жінок',female_dataset['Age'].max())

print('середній вік пасажирів',dataset['Age'].mean())
````
